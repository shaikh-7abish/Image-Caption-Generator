the list contains atleast 5 captions for a single image to save all captions to their respective image name, we need to store it in dictionary of lists.

```python
# creating function to create a dictionary and storing captions in it
def img_cap(filename):
  description = {}
  captions = filename
  for caption in captions:
    img,cap = caption.split('\t')
    img =  img.split('#')
    img = img[0]
    if img not in description:
      description[img] = [cap]
    else:
      description[img].append(cap)
  return description
```

```python
description = img_cap(texts)
```

```python
print('data type',type(description))
print('len', len(description))
print(description['1002674143_1b742ab4b8.jpg'])
```

update the keys to right name of images

```python
import string

def clean_text(file):
  table = str.maketrans('','',string.punctuation)
  newfile = {}
  for img, caps in file.items():
    for i, cap in enumerate(caps):
      cap.replace('-',' ')

      desc = cap.split()

      desc = [word.lower() for word in desc]

      desc = [word.translate(table) for word in desc]

      #remove hanging 's and a
      desc = [word for word in desc if(len(word)>1)]

      #remove tokens with numbers in them
      desc = [word for word in desc if(word.isalpha())]

      cap = ' '.join(desc)

      if i==0:
        newfile[img] = [cap]
      else:
        newfile[img].append(cap)

  return newfile
```

```python
cleaned_description = clean_text(description)
print('data type',type(cleaned_description))
print('len', len(cleaned_description))
```

```python
def save_descriptions(descriptions, filename):
    lines = list()
    for key, desc_list in descriptions.items():
        for desc in desc_list:
            lines.append(key + '\t' + desc )
    data = "\n".join(lines)
    file = open(filename,"w")
    file.write(data)
    file.close()
```

```python
#saving each description to file
save_descriptions(cleaned_description, "../files/descriptions.txt")
```
